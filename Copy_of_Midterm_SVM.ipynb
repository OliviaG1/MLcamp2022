# Feature: PoS

```python
pos_tags = [('CC', 'coordinating conjunction'), 
            ('IN', 'preposition/subordinating conjunction'), 
            ('JJ', 'adjective'),
            ('JJR', 'adjective, comparative'),
            ('JJS', 'adjective, superlative'),
            ('MD', 'modal'),
            ('NN', 'noun, singular'),
            ('NNS', 'noun plural'),
            ('NNP', 'proper noun, singular'),
            ('NNPS', 'proper noun, plural'),
            ('PRP', 'personal pronoun'),
            ('PRP$', 'possessive pronoun'),
            ('RB', 'adverb'),
            ('RBR', 'adverb, comparative'),
            ('RBS', 'adverb, superlative'),
            ('UH', 'interjection'),
            ('VB', 'verb, base form'),
            ('VBD', 'verb, past tense'),
            ('VBG', 'verb, gerund/present participle'),
            ('VBN', 'verb, past participle'),
            ('VBP', 'verb, sing. present, non-3d'),
            ('VBZ', 'verb, 3rd person sing. present')]

tag_names = []
for i in pos_tags:
  tag_names.append(i[0])

import spacy
def num_tags(document):
  tags = nltk.pos_tag(document)
  result=[]
  for i in pos_tags:
    current_tag=0
    for j in tags:
      if(j[1] == i[0]):
        current_tag += 1
    result.append(current_tag)
  return result
```

```python
from nltk.tokenize import word_tokenize
nltk.download('averaged_perceptron_tagger')

pos_train = []
for i in full_X_train[0]:
  pos_train.append(num_tags(word_tokenize(i)))

pos_test = []
for i in testing_data[0]:
  pos_test.append(num_tags(word_tokenize(i)))
```

```python
lengths_train = []
for i in full_X_train[0]:
  l = []
  l.append(len(i.split()))
  lengths_train.append(l)

lengths_test = []
for i in testing_data[0]:
  l = []
  l.append(len(i.split()))
  lengths_test.append(l)
```

```python
from nltk.stem import PorterStemmer
nltk.download('wordnet')
nltk.download('omw-1.4')
sp = spacy.load('en_core_web_sm')

def preprocessing(x):
  x = x.lower()
  all_stopwords = sp.Defaults.stop_words
  x = ' '.join([word for word in x.split() if word not in (all_stopwords)])
  x = ' '.join([word for word in x.split() if word.isalnum()])
  lemmatizer = nltk.stem.WordNetLemmatizer()
  x = ' '.join([lemmatizer.lemmatize(word) for word in x.split()])
  return(x)


full_X_train[0] = full_X_train[0].apply(preprocessing)
testing_data[0] = testing_data[0].apply(preprocessing)
```

# Feature: tf-idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
all_X_train = vectorizer.fit_transform(full_X_train[0])
```

```python
df_X_train = pd.DataFrame(all_X_train.todense())
df_X_train.columns = vectorizer.get_feature_names()
df_X_train

df_pos_train = pd.DataFrame(pos_train)
df_pos_train.columns = tag_names
df_pos_train

df_lengths_train = pd.DataFrame(lengths_train)
df_lengths_train.columns = ['Length']
df_lengths_train

X_train = pd.concat((df_X_train, df_pos_train, df_lengths_train), axis = 1)
X_train
```

# SVM

```python
all_X_test = vectorizer.transform(testing_data[0])

df_X_test = pd.DataFrame(all_X_test.todense())
df_X_test.columns = vectorizer.get_feature_names()

df_pos_test = pd.DataFrame(pos_test)
df_pos_test.columns = tag_names
df_pos_test

df_lengths_test = pd.DataFrame(lengths_test)
df_lengths_test.columns = ['Length']
df_lengths_test

X_test = pd.concat((df_X_test, df_pos_test, df_lengths_test), axis = 1)
X_test
```

```python
from sklearn import svm
clf = svm.SVC(C=10, kernel='linear')
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(X_test)
accuracy = accuracy_score(labels, y_pred)
accuracy
```
